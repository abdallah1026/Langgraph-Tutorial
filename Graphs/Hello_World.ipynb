```python
# Import Dependencies

from typing import TypedDict, Dict
from langgraph.graph import StateGraph
```

```python
# We Will create an AgentSatate class :- Shared data structure  that keep track of the information as you application run.ArithmeticError

class AgentSatate(TypedDict): # ---> our state schema

    message: str
```

```python
# Now we will create a Node


def greeting_node(state: AgentSatate) -> AgentSatate:
    """Simple Node that adding greeting to the state"""
    
    state["message"] = "Hello " + state["message"] + " How are you?"
    return state
```

```python
# Great, now let's create our graph

graph = StateGraph(AgentSatate)  # We Initialize the graph with our AgentState

# Now let's add our nodes to the graph

graph.add_node("greeting_node", greeting_node)

# Now let's set the entry point of our graph

graph.set_entry_point("greeting_node")

# Now let's set finish point of the graph.

graph.set_finish_point("greeting_node")

# Let's Compile the graph

app = graph.compile()
```

```python
# Let's test the graph

result = app.invoke({"message": "Abdallah"})
```
